```python
# for data analysis
import numpy as np
import pandas as pd

# for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('poster')
```

#### Creating new dataframe

```python
# m = 1
# c = -1

# x = np.arange(-10,11,1)
# print(x)

# y = m*x + c
# print(y)
```

```python
def straight_line(x, m, c):
    
    df = pd.DataFrame()
    
    df["x"] = x
    df["y"] = m*x + c
    
    return df
```

#### Creating new straight lines

```python
SL_1 = straight_line(np.arange(-10,11,1), 1, 1)
SL_2 = straight_line(np.arange(-10,11,1), -1, 1)
SL_3 = straight_line(np.arange(-10,11,1), -1, -1)
SL_4 = straight_line(np.arange(-10,11,1), 1, -1)
```

#### Create a lineplot function

```python
def lineplot(x_axis, y_axis, title, color):
    plt.plot(x_axis, y_axis, color=color, marker=".")
    plt.hlines(0, -10, 10, color="black", linestyles="-.")
    plt.vlines(0, -15, 15, color="black", linestyles=":")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.xticks(np.arange(-10,11,5))
    plt.yticks(np.arange(-15,16,5))
    plt.title(title)
```

#### Creating subplots

#### Varying the marker, color, and linestyle

```python
plt.figure(figsize=(20,16))

# position = 1, row = 1, col = 1
plt.subplot(3,3,1)
lineplot(SL_1["x"], SL_1["y"], "y=mx+c where m=1, c=1", "r")

# position = 2, row = 1, col = 2
plt.subplot(3,3,2)
lineplot(SL_2["x"], SL_2["y"], "y=mx+c where m=-1, c=1", "g")

# position = 3, row = 1, col = 3
plt.subplot(3,3,3)
lineplot(SL_3["x"], SL_3["y"], "y=mx+c where m=-1, c=-1", "orange")

# position = 4, row = 2, col = 1
plt.subplot(3,3,4)
lineplot(SL_4["x"], SL_4["y"], "y=mx+c where m=1, c=-1", "b")

plt.tight_layout()
plt.show()
```

#### Multiple lineplots in the same figure

```python
plt.figure(figsize=(20,12)) #(width, height)

plt.plot(SL_1['x'], SL_1['y'], color='red', label='m=1, c=1')
plt.plot(SL_2['x'], SL_2['y'], color='green', label='m=-1, c=1')
plt.plot(SL_3['x'], SL_3['y'], color='blue', label='m=-1, c=-1')
plt.plot(SL_4['x'], SL_4['y'], color='orange', label='m=1, c=-1')
plt.vlines(0, -15, 15, linestyles=':', colors="black")
plt.hlines(0, -10, 10, linestyles=':', colors="black")
plt.title("y=mx+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="center right") 
# plt.grid()
plt.show()
```
